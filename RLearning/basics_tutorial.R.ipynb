```r
t <- 1:365
temp.ideal <- 10 * sin(2*pi*(t - 100) / 365)
temp.noisy <- temp.ideal + rnorm(365, sd=3)

plot(t, temp.ideal, type='l', lwd=2, ylim=range(temp.noisy), )
points(temp.noisy, 
       pch=16,
       col=rgb(0.2, 0.3, 0.7, alpha=0.4))
lines(lowess(t, temp.ideal), col=2)
legend('topleft', 
       lty=1, 
       col=c(1, 2), 
       lwd=c(2, 1), 
       legend=c('Generating Model', 'Cubic Spline'))
```

# Control flow

```r
x <- numeric(50)
x[1] <- 1
for (i in 2:50) {
    x[i] <- 0.8*x[i - 1] + rnorm(1)
}

plot(x, type='l')
```

```r
x <- sample(10, 1)
if (x < 3) {
    cat(sprintf('%d < 3', x))
} else if (x < 6) {
    cat(sprintf('3  %d < 6', x))
} else {
    cat(sprintf('$d  6', x))
}
```

```r
target <- 10
i <- 1
x <- rpois(1, 3)
while (x < target) {
    i <- i + 1
    x <- rpois(1, 3)
}
sprintf('A value of %s or more occured after %d samples', target, i)
```

# Basic Linear Modeling

```r
data(Animals)
head(Animals)
```

```r
dim(Animals)
```

```r
plot(Animals$brain ~ Animals$body)
```

```r
plot(log(Animals$brain) ~ log(Animals$body))
```

```r
mod1 <- lm(brain ~ body, data=Animals)
summary(mod1)
```

```r
par(mfrow=c(2, 2))
plot(mod1)
```

```r
mod2 <- lm(log(brain) ~ log(body), data=Animals)
summary(mod2)
```

```r
par(mfrow=c(2, 2))
plot(mod2)
```

```r
plot(log(Animals$brain) ~ log(Animals$body))
text(log(Animals$body), log(Animals$brain), rownames(Animals))
abline(mod2, col=4)
```
